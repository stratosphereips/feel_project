```python
fed_acc_mean = [acc_day1_mean, acc_day2_mean, acc_day3_mean, acc_day4_mean]
fed_acc_std = [acc_day1_std, acc_day2_std, acc_day3_std, acc_day4_std]

fed_tpr_mean = [tpr_day1_mean, tpr_day2_mean, tpr_day3_mean, tpr_day4_mean]
fed_tpr_std = [tpr_day1_std, tpr_day2_std, tpr_day3_std, tpr_day4_std]

fed_fpr_mean = [fpr_day1_mean, fpr_day2_mean, fpr_day3_mean, fpr_day4_mean]
fed_fpr_std = [fpr_day1_std, fpr_day2_std, fpr_day3_std, fpr_day4_std]
```
