# FedAvgM(0.2)

```python
acc_day1 = [98.45, 97.43, 97.42, 97.44, 97.15]
tpr_day1 = [93.10, 93.79, 93.10, 97.24, 97.24]
fpr_day1 = [0.45, 1.81, 1.71, 2.47, 2.83]

acc_day1_mean = np.mean(acc_day1)
acc_day1_std = np.std(acc_day1)
tpr_day1_mean = np.mean(tpr_day1)
tpr_day1_std = np.std(tpr_day1)
fpr_day1_mean = np.mean(fpr_day1)
fpr_day1_std = np.std(fpr_day1)

acc_day2 = [97.57, 98.00, 98.01, 97.52, 98.56]
tpr_day2 = [93.10, 93.10, 93.10, 93.10, 93.1]
fpr_day2 = [1.49, 0.94, 0.92, 1.56, 0.29]

acc_day2_mean = np.mean(acc_day2)
acc_day2_std = np.std(acc_day2)
tpr_day2_mean = np.mean(tpr_day2)
tpr_day2_std = np.std(tpr_day2)
fpr_day2_mean = np.mean(fpr_day2)
fpr_day2_std = np.std(fpr_day2)

acc_day3 = [97.62, 98.22, 98.17, 98.10, 97.54]
tpr_day3 = [93.10, 97.24, 94.48, 97.24, 97.24]
fpr_day3 = [1.53, 2.75, 1.12, 1.76, 2.45]

acc_day3_mean = np.mean(acc_day3)
acc_day3_std = np.std(acc_day3)
tpr_day3_mean = np.mean(tpr_day3)
tpr_day3_std = np.std(tpr_day3)
fpr_day3_mean = np.mean(fpr_day3)
fpr_day3_std = np.std(fpr_day3)

acc_day4 = [96.97, 97.79, 97.35, 97.82, 97.05]
tpr_day4 = [97.24, 93.79, 98.62, 97.24, 97.24]
fpr_day4 = [3.08, 1.40, 2.88, 2.05, 2.99]

acc_day4_mean = np.mean(acc_day4)
acc_day4_std = np.std(acc_day4)
tpr_day4_mean = np.mean(tpr_day4)
tpr_day4_std = np.std(tpr_day4)
fpr_day4_mean = np.mean(fpr_day4)
fpr_day4_std = np.std(fpr_day4)

fed_avg_acc_mean = np.array([acc_day1_mean, acc_day2_mean, acc_day3_mean, acc_day4_mean])
fed_avg_acc_std = np.array([acc_day1_std, acc_day2_std, acc_day3_std, acc_day4_std])

fed_avg_tpr_mean = np.array([tpr_day1_mean, tpr_day2_mean, tpr_day3_mean, tpr_day4_mean])
fed_avg_tpr_std = np.array([tpr_day1_std, tpr_day2_std, tpr_day3_std, tpr_day4_std])

fed_avg_fpr_mean = np.array([fpr_day1_mean, fpr_day2_mean, fpr_day3_mean, fpr_day4_mean])
fed_avg_fpr_std = np.array([fpr_day1_std, fpr_day2_std, fpr_day3_std, fpr_day4_std])
```

```python
plt.plot(range(4), centr_acc_mean, marker='^', label="Centralized")
plt.fill_between(range(4), centr_acc_mean+centr_acc_std, centr_acc_mean-centr_acc_std, alpha=0.2)

plt.plot(range(4), fed_acc_mean, marker='^', label="FedAvg")
plt.fill_between(range(4), fed_acc_mean+fed_acc_std, fed_acc_mean-fed_acc_std, alpha=0.2)

plt.plot(range(4), fed_avg_acc_mean, marker='^', label="FedAvgM(0.2)")
plt.fill_between(range(4), fed_avg_acc_mean+fed_avg_acc_std, fed_avg_acc_mean-fed_avg_acc_std, alpha=0.2)


plt.legend(loc=4)

plt.ylim(80, 100.0)
plt.ylabel("Accuracy (%)")
plt.xlabel("Testing day")
plt.xticks(range(4), days)
# plt.title("Accuracy (%) per day")
plt.grid(axis='y')
```

```python
plt.plot(range(4), centr_tpr_mean, marker='^', label="Centralized")
plt.fill_between(range(4), centr_tpr_mean+centr_tpr_std, centr_tpr_mean-centr_tpr_std, alpha=0.2)

plt.plot(range(4), fed_tpr_mean, marker='^', label="FedAvg")
plt.fill_between(range(4), fed_tpr_mean+fed_tpr_std, fed_tpr_mean-fed_tpr_std, alpha=0.2)

plt.plot(range(4), fed_avg_tpr_mean, marker='^', label="FedAvgM(0.2)")
plt.fill_between(range(4), fed_avg_tpr_mean+fed_avg_tpr_std, fed_avg_tpr_mean-fed_avg_tpr_std, alpha=0.2)
plt.legend(loc=4)

plt.ylim(0, 100.0)
plt.ylabel("TPR (%)")
plt.xlabel("Testing day")
plt.xticks(range(4), days)
# plt.title("TPR (%) per day")
plt.grid(axis='y')
```
