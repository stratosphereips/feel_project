```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
centr_acc = [98.66, 98.71, 98.99, 98.95]
centr_tpr = [93.10, 93.10, 97.24, 97.24]
centr_fpr = [1.26, 1.09, 0.97, 1.42]
days = ["Day 2", "Day 3", "Day 4", "Day 5"]
```

```python
acc_day1 = [98.52, 98.3, 83.69, 96.68, 96.79]
tpr_day1 = [97.24, 93.1, 3.45, 93.79, 93.1]
fpr_day1 = [1.17, 0.63, 0.04, 2.7, 2.41]

acc_day1_mean = np.mean(acc_day1)
acc_day1_std = np.std(acc_day1)
tpr_day1_mean = np.mean(tpr_day1)
tpr_day1_std = np.std(tpr_day1)
fpr_day1_mean = np.mean(fpr_day1)
fpr_day1_std = np.std(fpr_day1)
```

```python
acc_day2 = [97.96, 97.76, 83.02, 88.91, 97.57]
tpr_day2 = [95.86, 93.1, 4.83, 37.93, 93.1]
fpr_day2 = [1.6, 1.25, 0.2, 0.14, 1.45]

acc_day2_mean = np.mean(acc_day2)
acc_day2_std = np.std(acc_day2)
tpr_day2_mean = np.mean(tpr_day2)
tpr_day2_std = np.std(tpr_day2)
fpr_day2_mean = np.mean(fpr_day2)
fpr_day2_std = np.std(fpr_day2)
```

```python
acc_day3 = [98.5, 98.01, 98.43, 97.98, 98.10]
tpr_day3 = [93.1, 93.10, 93.10, 94.48, 93.79]
fpr_day3 = [0.44, 1.01, 0.52, 1.32, 1.05]

acc_day3_mean = np.mean(acc_day3)
acc_day3_std = np.std(acc_day3)
tpr_day3_mean = np.mean(tpr_day3)
tpr_day3_std = np.std(tpr_day3)
fpr_day3_mean = np.mean(fpr_day3)
fpr_day3_std = np.std(fpr_day3)
```

```python
acc_day4 = [98.05, 97.79, 97.76, 98.12, 97.85]
tpr_day4 = [97.24, 93.79, 97.24, 93.10, 97.24]
fpr_day4 = [1.75, 1.40, 2.11, 0.88, 2.00]

acc_day4_mean = np.mean(acc_day4)
acc_day4_std = np.std(acc_day4)
tpr_day4_mean = np.mean(tpr_day4)
tpr_day4_std = np.std(tpr_day4)
fpr_day4_mean = np.mean(fpr_day4)
fpr_day4_std = np.std(fpr_day4)
```

```python
fed_acc_mean = [acc_day1_mean, acc_day2_mean, acc_day3_mean, acc_day4_mean]
fed_acc_std = [acc_day1_std, acc_day2_std, acc_day3_std, acc_day4_std]

fed_tpr_mean = [tpr_day1_mean, tpr_day2_mean, tpr_day3_mean, tpr_day4_mean]
fed_tpr_std = [tpr_day1_std, tpr_day2_std, tpr_day3_std, tpr_day4_std]

fed_fpr_mean = [fpr_day1_mean, fpr_day2_mean, fpr_day3_mean, fpr_day4_mean]
fed_fpr_std = [fpr_day1_std, fpr_day2_std, fpr_day3_std, fpr_day4_std]
```

```python
plt.plot(centr_acc, marker="x", label="centralized")
plt.errorbar(range(4), fed_acc_mean, yerr=fed_acc_std, fmt='o-', color='orange',
             ecolor='lightgray', elinewidth=3, capsize=0, label="federated")
plt.legend(loc=4)

plt.ylim(50, 100.0)
plt.ylabel("Accuracy (%)")
plt.xticks(range(4), days)
plt.title("Accuracy (%) per day")
plt.grid(axis='y')
```

```python
plt.plot(centr_tpr, marker="x", label="centralized")
plt.errorbar(range(4), fed_tpr_mean, yerr=fed_tpr_std, fmt='o-', color='orange',
             ecolor='lightgray', elinewidth=3, capsize=0, label="federated")
plt.legend(loc=4)

plt.ylim(0, 100.0)
plt.ylabel("TPR (%")
plt.xticks(range(4), days)
plt.title("TPR (%) per day")
plt.grid(axis='y')
```
