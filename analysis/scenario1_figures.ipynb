# Triplet loss

```python
acc_day1 = [98.29, 98.14, 98.15, 97.87, 96.98]
tpr_day1 = [97.24, 97.24, 93.10, 97.24, 93.1]
fpr_day1 = [1.48, 1.66, 0.84, 1.99, 2.28]

acc_day1_mean = np.mean(acc_day1)
acc_day1_std = np.std(acc_day1)
tpr_day1_mean = np.mean(tpr_day1)
tpr_day1_std = np.std(tpr_day1)
fpr_day1_mean = np.mean(fpr_day1)
fpr_day1_std = np.std(fpr_day1)

acc_day2 = [98.55, 97.95, 98.18, 98.28, 98.56]
tpr_day2 = [97.24, 97.24, 93.10, 93.10, 93.10]
fpr_day2 = [1.19, 1.91, 0.76, 0.63, 0.29]

acc_day2_mean = np.mean(acc_day2)
acc_day2_std = np.std(acc_day2)
tpr_day2_mean = np.mean(tpr_day2)
tpr_day2_std = np.std(tpr_day2)
fpr_day2_mean = np.mean(fpr_day2)
fpr_day2_std = np.std(fpr_day2)

acc_day3 = [98.14, 98.23, 98.77, 98.67, 97.73]
tpr_day3 = [93.79, 97.24, 97.24, 96.55, 93.10]
fpr_day3 = [1.0, 1.59, 0.96, 0.92, 1.41]

acc_day3_mean = np.mean(acc_day3)
acc_day3_std = np.std(acc_day3)
tpr_day3_mean = np.mean(tpr_day3)
tpr_day3_std = np.std(tpr_day3)
fpr_day3_mean = np.mean(fpr_day3)
fpr_day3_std = np.std(fpr_day3)

acc_day4 = [97.24, 97.47, 97.00, 97.54, 96.98]
tpr_day4 = [97.24, 97.24, 97.24, 97.24, 94.48]
fpr_day4 = [2.76, 2.49, 3.05, 2.4, 2.54]

acc_day4_mean = np.mean(acc_day4)
acc_day4_std = np.std(acc_day4)
tpr_day4_mean = np.mean(tpr_day4)
tpr_day4_std = np.std(tpr_day4)
fpr_day4_mean = np.mean(fpr_day4)
fpr_day4_std = np.std(fpr_day4)

triplet_acc_mean = np.array([acc_day1_mean, acc_day2_mean, acc_day3_mean, acc_day4_mean])
triplet_acc_std = np.array([acc_day1_std, acc_day2_std, acc_day3_std, acc_day4_std])

triplet_tpr_mean = np.array([tpr_day1_mean, tpr_day2_mean, tpr_day3_mean, tpr_day4_mean])
triplet_tpr_std = np.array([tpr_day1_std, tpr_day2_std, tpr_day3_std, tpr_day4_std])

triplet_fpr_mean = np.array([fpr_day1_mean, fpr_day2_mean, fpr_day3_mean, fpr_day4_mean])
triplet_fpr_std = np.array([fpr_day1_std, fpr_day2_std, fpr_day3_std, fpr_day4_std])
```

```python
plt.plot(range(4), centr_acc_mean, marker='^', label="Centralized")
plt.fill_between(range(4), centr_acc_mean+centr_acc_std, centr_acc_mean-centr_acc_std, alpha=0.2)

plt.plot(range(4), fed_acc_mean, marker='^', label="FedAvg")
plt.fill_between(range(4), fed_acc_mean+fed_acc_std, fed_acc_mean-fed_acc_std, alpha=0.2)

plt.plot(range(4), fed_avg_acc_mean, marker='^', label="FedAvgM(0.2)")
plt.fill_between(range(4), fed_avg_acc_mean+fed_avg_acc_std, fed_avg_acc_mean-fed_avg_acc_std, alpha=0.2)

plt.plot(range(4), triplet_acc_mean, marker='^', label="Triplet")
plt.fill_between(range(4), triplet_acc_mean+triplet_acc_std, triplet_acc_mean-triplet_acc_std, alpha=0.2)

plt.legend(loc=4)

plt.ylim(80, 100.0)
plt.ylabel("Accuracy (%)")
plt.xlabel("Testing day")
plt.xticks(range(4), days)
# plt.title("Accuracy (%) per day")
plt.grid(axis='y')
```
